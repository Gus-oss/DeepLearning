# Neuronas Artificiales

## Introducción

Los perceptrones se inventaron como modelos computacionales simples de neuronas. Una sola neurona suele tener múltiples entradas (dendritas), un cuerpo celular y una única salida (el axón).

image.png

Las variables son las $x_{1} , x_{2}, x_{3}, ... , x_{n}$ y los pesos son $w_{1}, w_{2}, w_{3},...w_{n}$ (Las variables nolas podemos cambiar, podemos cambiar los pesos). Posteriormente se forma, $z = \sum_{i=1}^{n}x_{i}w_{i} + b$ donde b es el bias (intercepto) (en la imagen esta es la variable $x_{0}$ y tiene peso 1).Posteriormente, introducimos no linealidad en el resultado, aplicando una función llamada función de activación o de transferencia. Por lo tanto, nuestra salida es $\hat{y} = f(z)$ (En lo que la imagen es Output es $\hat{y}$).

En resumen,  una neurona toma la entrada, $x$, la multiplica por pesos, $w$, y le suma el sesgo, $b$ , y luego se le aplica la función de activación y obtenemos la salida.

```python
# Este es un ejemplo de un perceptron de 3 dentritas , 1 cuerpo y 1 axon
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def neurona1(x1,x2,x3): #dentritas
  w1 =0.25 #cuerpo
  w2 = 0.4 #cuerpo
  w3 = 0.35 #cuerpo
  bias =1.25 #cuerpo
  z = w1*x1 +w2*x2 +w3*x3 +bias #cuerpo
  a1 = max(0,z) #axon
  return a1
# este es un ejemplo de propagacion hacia adelante y esto ya es inteligencia artificial, por que simula el comportamiento de una neurona

neurona1(1,5,6)
```

## Neurona artificial - propagacion hacia adelante (Forward Propagation)

La propagación hacia adelante (también conocida como pase hacia adelante) es el proceso mediante el cual una neurona artificial (o una red completa) transforma los datos de entrada en una predicción de salida. Es la fase de "pensamiento" de la neurona, donde la información fluye en una sola dirección: desde las entradas, pasando por las operaciones matemáticas, hasta la estimación final (Un ejemplo de propagacion hacia adelante aplicado es el ejemplo de neuroa que se programo anteriormente).

## Neuronas artificiales - propagacion hacia atras (Backward propagation)

La propagación hacia atrás (o retropropagación) responde a la pregunta ¿Como debo cambiar los pesos para reducir el error?. Mientras que la propagación hacia adelante calcula una predicción, la retropropagación va en sentido inverso para determinar con exactitud cuánto contribuyó cada ponderación y sesgo al error de predicción.

image.png

Definimos una nueva función llamada función de costo (J), también conocida como función de pérdida (L), que nos indica el rendimiento de nuestra red neuronal (La idea centras es aplicar propagacion hacidelante, despues utilizar la funcion de costo para indicar el rendimiento de la prediccion y ajustar los pesos y volver a hacer la propagacion hacia adelante). Si el costo o la pérdida es muy alto, significa que nuestra red no está prediciendo el resultado correcto. Por lo tanto, nuestro objetivo es minimizar la función de costo para que las predicciones de nuestra red neuronal sean mejores. ¿Cómo podemos minimizar la función de costo? Es decir, ¿cómo podemos minimizar la pérdida/costo? Aprendimos que la red neuronal realiza predicciones mediante propagación hacia adelante. Por lo tanto, si podemos cambiar algunos valores en la propagación hacia adelante, podemos predecir el resultado correcto y minimizar la pérdida.

Veamos unos ejemplos de funciones de perdida

### Funciones de perdida.

#### Error cuadratico medio [Mean Squared Error- MSE]

Esta función se define como la media de la diferencia al cuadrado entre la salida real y la salida predicha. Se utiliza cuando se desea que la red neuronal prediga un valor escalar continuo.

$L(\theta) = \frac{1}{N} \Sigma_{i=1}^{N}(y_{i}-\hat{y}_{i})^{2}$

#### Error absoluto medio [Mean Absolute Error- MAE]

Esta función se define como el valor absoluto de la diferencia entre la salida real y la salida predicha. Se utilzia cuando hay mucha volatilidad en los datos.

$L(\theta) = \frac{1}{N} \Sigma_{i=1}^{N} |y_{i}-\hat{y}_{i}|$

#### Entropia cruzada (Cross- Entropy)

Siempre se debe usar  la pérdida de entropía cruzada si hay probabilidades involucradas. Es decir, si está realizando algún tipo de clasificación.

$L(\theta) = - \Sigma_{i=1}^{N} y_{i} \cdot log(\hat{y}_{i})$

#### Error absoluto medio porcentual (Mean Absolute Percentage Error- MEAPE)

La función de pérdida de error porcentual absoluto medio debe utilizarse durante las tareas de pronóstico de la demanda para evaluar el rendimiento del modelo durante el entrenamiento.

$MAPE = \frac{100%}{N}\Sigma_{i=0}^{N}\frac{|y_{i}-\hat{y}_{i}|}{\hat{y}_{i}}$

### Decenso por Gradiente

El decenso por gradiente es uno de los algoritmos mas usados en optimización de algoritmos. Se utiliza para minimizar la función de costos (min $L$) que a su vez minimiza los errores y obtiene el error mas bajo posible.

Habiamos comentado que los pesos eran la clave para un buen algoritmo, ahora nos podemos preguntar ¿Como podemos encontrar los pesos optimos por el desenso por gradiente?

En el siguiente grafico de los pesos vs la funcion de costos nos muestra como seria la solución. Nuestro objetivo seria minimizar la funcion de costos, partiendo desde unos pesos iniciales y llegando el punto hasta donde el costo sea minimo.

image.png

Ahora surge la duda, ¿Como movemos el punto hacia abajo? (es decir, ¿Como encontramos los pesos tales que minimizen el costo?) ¿Como los buscamos?. Para esto, utilizamos el gradiente. La idea es calcular el gradiente de los pesos iniciales e ir calculado el gradiente de la funcion de costos respecto a los pesos ($\frac{\partial J}{\partial W}$)hasta encontrar su minimo con respecto a los pesos iniciales (Ver la siguiente figura).


image.png

El desencso por gradiente es un algoritmo de optimizacion de primer orden, es decir, solo considera las primeras derivadas.


Con el algoritmo de desenso por gradiente podemos mover nuestros pesos iniciales hasta un peso que minimize la funcion de costos, pero ¿Como actualizamos los pesos? ¡Con el Output de la propagación hacia adelante! La idea es, partir de nuestros pesos iniciales, formar $z$ y pasar esta $z$ a la función de activación, posteriormente obtener nuestra $\hat{y}$, obtenenemos el gradiente respecto a los pesos y actualizamos nuestros pesos por medio de:

$W_{nuevos} = W_{anteriores} - \alpha\frac{\partial J}{\partial W_{anteriores}}$


Aqui $\alpha$ se le llama taza de aprendizaje y usualmente es un numero muy pequeño (entre 0 y 1, aun que casi nunca se pone 1, es mas comun poner 0.1). Esta taza de aprendizaje me modula la magnitud del $\frac{\partial J}{\partial W_{anteriores}}$, es decir, me indica que tanto me voy a mover. Si no tuvieramos este valor tendriamos saltos muy grandes al buscar el minimo (o con movimientos muy bruzcos, ver imagen siguiente). Tambien modula la diferencia entre unidades que hay entre W y $\frac{\partial J}{\partial W_{anteriores}}$.

### Decesno por Gradiente con momento

El desenso por gradiente con momento es una tecnica de optimizacion que acelera la convergencia del decesno por gradiente, especialmente en escenarios donde el proceso de optimizacion es lento o se trunca en un minimo local (es decir, no se alcance el minimo global debido a la falta de datos).Se introduce el concepto de momentum o momento, el cual es un parametro que actualiza (con ayuda del algorimto) mas eficientemente a lo largo de las direcciones la solucion optima.

image.png

Es decir, si parto en A y me quedo en B por falta d edatos, se implementa se añade el Momentum para alzanzar el minimo local.


La idea es incorporar el concepto de "velocidad" a la formula

$\theta_{actual} = \theta_{anterior} - \eta \frac{\partial J(\theta)}{\partial \theta}$

**Reglas de actualización**

- actualización de la velocidad: $v = γ \cdot v + \eta \frac{\partial J(\theta)}{\partial \theta}$


donde v es la velocidad (se inicializa en 0), $\gamma$ es el coeficiente del momentum , el cual es un iperparametro de denota la influecnia del pasado gradiente (usualmente oma los valores entre 0-1), $\eta$ es la regla de aprendizaje y $\frac{\partial J(\theta)}{\partial \theta}$ es el gradiente de la funcion de costos respecto a los parametros.
- Actualziacion de los parametros: Los parametros se actualizan mediante: $\theta_{actual} = \theta_{anterior} - v$
