### Ejercicio - Variables

```python
#Leer Tabla de Datos
DB = pd.read_csv("C:/Users/Mauricio/Downloads/EarlyRetirementPrediction.csv") # Aqui va donde ustedes guarden el archivo
DB = DB.drop('ClienteID',axis=1) #Elimina el ClienteID del Set de Datos
DB
```

```python
#Seleccion de variables
X = DB.drop('Retire Before 65 Years Old',axis=1) #Seleccionamos 3 Variables Predictivas
Y = DB[['Retire Before 65 Years Old']] #Seleccionamos la variable Objetivo
```

```python
#Escalamos nuestras variables
X_Scaled = StandardScaler()
Y_Scaled = StandardScaler()
X = X_Scaled.fit_transform(X)
Y = Y_Scaled.fit_transform(Y)
```

## Modelo Sigmoide

```python
def sigmoid(x):
    return 1/ (1+np.exp(-x))

def sigmoidPrime(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

## Definicion de la Neurona

```python

#Se define una neurona que procesa 3 inputs con 3 pesos para propagaciÃ³n hacia adelante 
# y el gradiente estocastico para propagaciÃ³n hacia atrÃ¡s
eps=1e-8 #eps sustituye si obtenemos un log negativo ya que el minimo es 0
seed=42
rng = np.random.default_rng(seed) #Pongo por default me semilla para que siempre use los mismos nÃºmero aleatorios

lr=0.01 #lr es tasa de aprendizaje; 
N, d = X.shape #N es mi nÃºmero de filas y d mi nÃºmero de columnas

# InicializaciÃ³n aleatoria de pesos
w = rng.random(d) #Esto genera una los 3 pesos de manera aleatoria
history = []#Guarda los errores a travez de las iteraciones

for i in range(N): #Iteramos sobre cada fila o X
    # Forward - PropagaciÃ³n hacia adelante
    z = X[i] @ w #Esto no es mÃ¡s que una multiplicaciÃ³nnde matrices
    y_hat = sigmoid(z) #Aplica la sigmoide

    # Loss (MSE)
    error = Y[i] - y_hat # Calcula el error de la estimaciÃ³n
    J = np.mean(error**2) #Eleva al cuadrado el error para penalizarlo
    history.append(J) #Almacena el resultado


    # Gradiente (estocastico)
    grad = (Y[i] - y_hat)*sigmoidPrime(z)*X[i] #Gradiente de la Sigmoide
    # Update
    w = w - lr * grad #Calcula los nuevo Pesos
    print(w,J) #Devuelve el peso ideal y mis resultados historicos
```
