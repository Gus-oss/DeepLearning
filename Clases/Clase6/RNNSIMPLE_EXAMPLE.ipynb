```python
#Supongamos que tenemos una red neuronal recurrent con dos neuronas en la capa oculta
#Initialization the weights

Ux =  [[ 0.18662322, 1.2369459 ]]  
Wh =  [[ 0.86981213, 0.49338293],
       [ 0.49338293,  0.8698122 ]]  
bh =  [0, 0]  
Vy = [[0.4635998],
      [0.6538409]] 
by =  [0]
```

```python
#Recurrent process with RELU as activation function
m=2 #Two recurrent neurons 
h0 = np.zeros(m)
h1 = np.dot(x[0], Ux) + h0 + bh #x[0]=31.42 
h2 = np.dot(x[1], Ux) + np.dot(h1,Wh) + bh #x[1]=32.77
h3 = np.dot(x[2], Ux) + np.dot(h2,Wh) + bh #x[2]=32.64
h4 = np.dot(x[3], Ux) + np.dot(h3,Wh) + bh#x[3]=33.95
y4 = np.dot(h4, Vy) + by #y4=f(h4(h3(h2(h1(h0)))))
```
